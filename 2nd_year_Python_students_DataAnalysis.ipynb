### グラフのある点を目立たせる

```python
# sns.set_context('talk')
df.plot.scatter(x='身長', y='体重', c='b')

x = df.iloc[3]['身長']
y = df.iloc[3]['体重']
plt.plot(x, y, c='r', marker='X', markersize=15)

plt.tick_params(labelsize=12) # 数値のフォントサイズの変更
plt.title('私はどこにいるか？')
plt.show()
```

> 1行目の `sns.set_context('talk')` は、`seaborn` の描画設定を変更して、グラフのフォントサイズや要素の大きさを「トーク（プレゼンテーション）」向けに調整するためのものです。具体的には、グラフの軸ラベル、タイトル、数値ラベルなどが大きめに設定され、視覚的にわかりやすいグラフになります。

`'talk'` は以下の4つのプリセットの1つです：
1. `'paper'`（デフォルト）: 小さめのフォントサイズ、通常のレポートや文書に適した設定。
2. `'notebook'`: ノートブック向け、やや大きめ。
3. `'talk'`: プレゼンテーションや口頭発表向けの大きめ設定。
4. `'poster'`: ポスターや展示用の非常に大きい設定。

この設定によって、`matplotlib` の標準フォントサイズなどが調整されます。



### グラフに線を引く

```python
# sns.set_context('paper')
df.plot.scatter(x='身長', y='体重', c='b')

x = df.iloc[3]['身長']
y = df.iloc[3]['体重']
plt.plot(x, y, c='r', marker='X', markersize=15)

plt.tick_params(labelsize=12) # 数値のフォントサイズの変更
plt.title('私はどこにいるか？')

plt.axvline(x=x, c='r', linestyle='--') # 縦線
plt.axhline(y=y, c='r', linestyle='--') # 横線

plt.show()
```

```python
%matplotlib inline
import japanize_matplotlib
import random
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font='IPAexGothic')

# ゴルトンボード表示関数：段数、玉数を指定する
def galton(steps, count):
    # 玉が落ちた位置を入れる空のリスト
    ans = []
    # 指定された玉数だけ繰り返す
    for i in range(count):
        # 玉を落とす最初の位置を50にする
        val = 50
        # 指定された段数だけ繰り返す
        for j in range(steps):
            # 0か1のランダムで、0なら-1, 1なら+1
            if random.randint(0,1) == 0:
                val -= 1
            else:
                val += 1
        # 最終的に玉が落ちた位置をリストに追加する
        ans.append(val)
    # 落下した結果のリストをデータフレームにして
    df = pd.DataFrame(ans)
    # 0列目(落とした結果の列)をヒストグラム表示
    df[0].plot.hist()
    plt.title(str(steps)+'段:'+str(count)+"個")
    plt.ylabel('')
    plt.show()
```

```python
galton(1,10000)
```

```python
galton(2,10000)
```

```python
galton(6,10000)
galton(10,100000)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    '数学': [100,85,90,95,80,80,75,65,65,60,55,45,45],
    '理科': [94,90,95,90,85,80,75,70,60,60,50,50,48],
    '社会': [80,88,70,62,86,70,79,65,75,67,75,68,60]
}
df = pd.DataFrame(data)

# Use combinations to generate pairs of columns without duplicates
from itertools import combinations
for i, j in combinations(df.columns, 2):
    df.plot.scatter(x=i, y=j, c='b')
    plt.title(f'{i}と{j}の関係性')
    plt.show()
```
